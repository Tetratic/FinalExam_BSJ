```python
import prettytable as pt
import math
```

```python
def cylinderVolume(r,h):
    '''Calculates volume of cylinder based on inputted radius and height'''
    v = (math.pi) * (r ** 2) * h
    v = format(v,'.2f')
    return v
```

```python
def cylinderSurface(r,h):
    '''Calculates surface area of cylinder based on inputted radius and height'''
    sA = 2 * (math.pi) * r * h + 2 * (math.pi) * (r ** 2)
    sA = format(sA,'.2f')
    return sA
```

```python
def coneVolume(r,h):
    '''Calculates volume of cone based on inputted radius and height'''
    v = (math.pi) * (r ** 2) * (h/3)
    v = format(v,'.2f')
    return v
```

```python
def coneSurface(r,h):
    '''Calculates surface area of cone based on inputted radius and height'''
    sA = (math.pi) * r * (r + math.sqrt((h ** 2) + (r ** 2)))
    sA = format(sA,'.2f')
    return sA
```

```python
def display_menu():
    '''Displays main menu to select an option from every loop'''
    print("     Geometry Calculator     ")
    print("=============================")
    print("1. Calculate the volume of a cylinder")
    print("2. Calculate the surface area of a cylinder")
    print("3. Calculate the volume of a cone")
    print("4. Calculate the surface area of a cone")
    print("5. Print Table")
    print("6. Save Data")
```

```python
def print_table(data):
    '''Displays collected data from every calculation performed since start of program'''
    x = pt.PrettyTable()
    
    x.field_names = ["Measurement","Radius","Height","Value"]
    
    for i in data:
        x.add_row(i)
    x.align = "c"
    x.align["Measurement"] = "r"
    print(x)
```

```python
def save_data(fn,data):
    '''Writes collected data to a file of the users choosing, including a header line to denote what each value is'''
    textline = ""
    
    x = open(fn,"w+")
    x.write("Measurement, Radius, Height, Value \n\n")
    for i in data:
        textline = ""
        for j in range(0,4):
            textline = textline + ", " + str(i[j])
        
        x.write(textline[2:] + "\n")
    x.close()
```

```python
def main():
    '''Runs main loop of program: Loops through the six options until the user wishes to exit'''
    print("Welcome to Geometry Calculator!")
    opt = 0
    
    outList = []
    
    while opt != -1:
        
        display_menu()
        opt = int(input("Choose option (-1 to quit): "))
        print("\n")
        
        if opt == 1:
            rad = int(input("Please enter the radius of the cylinder: "))
            hei = int(input("Please enter the height of the cylinder: "))
            vol = cylinderVolume(rad,hei)
            
            print("The volume of the cylinder is " + vol + " cubic units.")
            
            outList.append(["Cylinder Volume",rad,hei,vol])
            
        elif opt == 2:
            rad = int(input("Please enter the radius of the cylinder: "))
            hei = int(input("Please enter the height of the cylinder: "))
            surfA = cylinderSurface(rad,hei)
            
            print("The surface area of the cylinder is " + surfA + " square units.")
            
            outList.append(["Cylinder Surface Area",rad,hei,surfA])
        
        elif opt == 3:
            rad = int(input("Please enter the radius of the cone: "))
            hei = int(input("Please enter the height of the cone: "))
            vol = coneVolume(rad,hei)
            
            print("The volume of the cone is " + vol + " cubic units.")
            
            outList.append(["Cone Volume",rad,hei,vol])
            
        elif opt == 4:
            rad = int(input("Please enter the radius of the cone: "))
            hei = int(input("Please enter the height of the cone: "))
            surfA = coneSurface(rad,hei)
            
            print("The surface area of the cone is " + surfA + " square units.")
            
            outList.append(["Cone Surface Area",rad,hei,surfA])
            
        elif opt == 5:
            print_table(outList)
            
        elif opt == 6:
            filename = input("Please enter the filename you wish to save your data to, including the file format(.txt, .csv, etc.): ")
            
            save_data(filename, outList)
        
        elif opt == -1:
            print("Thank you for using Geometry Calculator!")
            break
        else:
            print("Invalid Option, please try again.")
            
    
```

```python
main()
```
